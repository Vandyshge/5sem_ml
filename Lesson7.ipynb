## Импортируем нужные библиотеки

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import regularizers
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

### Нейронная сеть для линейной функции

```python
X = np.array([1,2,3,4,5]).reshape(5,1)
y = np.array([1,2,3,4,5])
```

```python
# определение модели. В данном случае выбирается входной слой, 
# далее один скрытый с 10 нейронами, и один выходной с 1 нейроном, в качестве активационной функции используется ReLU
model = keras.models.Sequential([
    keras.layers.Dense(10, activation="relu", input_shape=X.shape[1:]),
    keras.layers.Dense(10, activation="relu", kernel_regularizer=regularizers.l2(0.1)),
    keras.layers.Dense(1)
])
```

```python
#определение функции потерь (которая минимизируется) и спопоба ее минимихации.
model.compile(loss="mean_squared_error", optimizer=keras.optimizers.SGD(lr=1e-3))
#обучение нейронной сети, определяется количество циклов обучения и валидационная выборка
model.fit(X, y, epochs=1000)
```

```python
model.get_weights()
```

```python
# Определим тестовые данные

X_test = np.linspace(1,100,num=100)
y_test = X_test
```

```python
fig = plt.figure(figsize=(8,5))
plt.plot(X_test, y_test, '--k')
plt.plot(model.predict(X_test), y_test, '-r', linewidth=3)
plt.xlabel('Predicted')
plt.ylabel('Real')
plt.show()
```

### Нейронная сеть для квадратичной

```python
# Теперь зададим квадратичную функцию
np.random.seed(42)
N = 100
x = np.linspace(-5.0, 5.0, N)
y = x**2 + np.random.normal(scale=2.5, size=N)
plt.scatter(x,y)
x = x.reshape(N,1)
plt.show()
```

```python
#разбиение данных на обучающую, валидационную и тестовую выборки.
X_train_full, X_test, y_train_full, y_test = train_test_split(x, y, shuffle=True, test_size=0.2)
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full, shuffle=True, test_size=0.2)

#нормировка данных, так как далее будет использоваться градиентный спуск
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_valid = scaler.transform(X_valid)
X_test = scaler.transform(X_test)
```

```python
# определение модели

model = keras.models.Sequential([
    keras.layers.Dense(10, activation="relu", input_shape=X_train.shape[1:]),
    keras.layers.Dense(5, activation="relu"),
    keras.layers.Dense(2, activation="relu"),
    keras.layers.Dense(10, activation="relu"),
    keras.layers.Dense(1)
])
```

```python
#определение функции потерь (которая минимизируется) и спопоба ее минимихации.
model.compile(loss="mean_squared_error", optimizer=keras.optimizers.SGD(lr=1e-3))

#обучение нейронной сети, определяется количество циклов обучения и валидационная выборка
history = model.fit(X_train, y_train, epochs=1000, validation_data=(X_valid, y_valid), verbose=2)
```

```python
plt.figure(figsize=(10,7))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'valid'], loc='upper left')
plt.show()
```

```python
y_pred = model.predict(X_train)
plt.scatter(X_train,y_train)
plt.plot(X_train, y_pred, 'ok')
plt.show()
```

```python
y_pred = model.predict(X_valid)
plt.scatter(X_valid,y_valid)
plt.plot(X_valid, y_pred, 'ok')
plt.show()
```

```python
y_pred = model.predict(X_test)
plt.scatter(X_test,y_test)
plt.plot(X_test, y_pred, 'ok')
plt.show()
```

```python
#этими командами можно сохранять обученную сеть и потом подгружать. По сути, сохраняется архитектура сети и значения весов
model.save("my_keras_model")
model = keras.models.load_model("my_keras_model")
```

## Домашнее задание

Обучить нейронную сеть для предсказания предела текучести (0.2% Proof Stress (MPa)) сталей на основании состава и температуры. Для этого подобрать наилучшим образом ее гиперпаметры (количество слоев, количество нейронов в слое, функции активации) построить зависимость предсказанного от реального значения и рассчитать ошибку (всё для обучающей и тестовой выборок). Сравнить точность с ранее обученными моделями

```python
# !pip install tensorflow
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import regularizers
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, ParameterGrid
from sklearn.preprocessing import StandardScaler
import pandas as pd
from sklearn.metrics import mean_squared_error as mse
```

```python
df = pd.read_csv('Steels_kaggle.csv')
y = df[' 0.2% Proof Stress (MPa)']
X = df[[' C', ' Si', ' Mn', ' P', ' S', ' Ni', ' Cr', ' Mo', ' Cu', 'V', ' Al', ' N', 'Ceq', 'Nb + Ta', ' Temperature (°C)']]
scaler = StandardScaler()
X_scaler = scaler.fit_transform(X)
X_train, X_test_, y_train, y_test_ = train_test_split(X_scaler, y, test_size=0.2, random_state=43)
X_test, X_valid, y_test, y_valid = train_test_split(X_test_, y_test_, test_size=0.2, random_state=43)

y_train, y_test, y_valid = y_train.ravel(), y_test.ravel(), y_valid.ravel()
```

```python
model = keras.models.Sequential([
    keras.Input(shape=X_train.shape[1:]),
    keras.layers.Dense(2, activation="relu"),
    keras.layers.Dense(4, activation="relu"),
    # keras.layers.Dense(5, activation="relu"),
    # keras.layers.Dense(10, activation="relu"),
    keras.layers.Dense(1)
])
```

```python
model.compile(loss="mean_squared_error", optimizer=keras.optimizers.SGD(lr=1e-3))

history = model.fit(X_train, y_train, epochs=500, validation_data=(X_valid, y_valid), verbose=2)
```

```python
plt.figure(figsize=(10,7))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.yscale('log')
plt.legend(['train', 'valid'], loc='upper left')
plt.show()
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
y_valid_pred = model.predict(X_valid)
# print(y_test_pred)

fig = plt.figure(figsize=(8,6))
plt.plot(y_train, y_train_pred, '.r', label='Train')
plt.plot(y_test, y_test_pred, '.b', label='Test')
plt.plot(y_valid, y_valid_pred, '.g', label='Valid')
plt.xlabel('Real value')
plt.ylabel('Predicted value')
plt.legend()
plt.show()

print(f'RMSE for nn (test) {mse(y_test, y_test_pred, squared=False)}')
```

```python
def grid_research(params):
  grid = ParameterGrid(params)
  loss = []
  for i in range(len(grid)):
    params_ = grid[i]
    print(i, params_)
    model = keras.models.Sequential()
    model.add(keras.Input(shape=X_train.shape[1:]))
    for _ in range(params_['n_layers']):
      model.add(keras.layers.Dense(params_['n_neurons'], activation=params_['activation']))
    model.add(keras.layers.Dense(1))

    model.compile(loss="mean_squared_error", optimizer=keras.optimizers.SGD(lr=params_['lr']))

    history = model.fit(X_train, y_train, epochs=200, validation_data=(X_valid, y_valid), verbose=2)
    # tmp = np.mean(history.history['val_loss'][50:])
    tmp = np.min(history.history['val_loss'])
    loss.append([tmp, i])
  
  loss = np.array(loss)
  np.savetxt('loss.txt', loss)
  return [grid[np.argmin(loss[:, 0])], np.min(loss[:, 0])]
                        
```

```python
# params = {
#     'n_layers': [1, 2, 3],
#     'n_neurons': [2, 3, 5],
#     'activation': ['relu', 'sigmoid', 'tanh'],
#     'lr': [1e-3, 1e-2]
# }

# params = {
#     'n_layers': [1, 2],
#     'n_neurons': [2, 3, 5],
#     'activation': ['relu', 'sigmoid', 'tanh'],
#     'lr': [1e-3, 1e-2]
# }

params = {
    'n_layers': [1, 2, 3],
    'n_neurons': [2, 3, 4],
    'activation': ['relu', 'sigmoid', 'tanh'],
    'lr': [1e-3]
}

out = grid_research(params)
```

Получил {'n_neurons': 4, 'n_layers': 2, 'lr': 0.001, 'activation': 'sigmoid'}, min loss validation = 976. При этом на RMSD_test = 41, что намного хуже чем было до этого
